```python
import pandas as pd

df = pd.read_json('orders_simple.json', orient="records")
df.head()
```

```python

df['total_amount'] = df['total_amount'].str.replace('$', '', regex=False)
df = df.astype({"total_amount": "float","shipping_days": "int64","customer_age": "int64"})
df["order_date"] = pd.to_datetime(df["order_date"])
df.head()


```

```python

df['items_html'] = df['items_html'].str.replace('<[^<]+?>', ' ', regex=True)
df.head()
```

```python
#3
df['coupon_used'] = df['coupon_used'].replace('','no coupon').fillna('no coupon')
```

```python
#4
df['order_month']=df['order_date'].dt.month
df.head()
```

```python
#5

# average_amount = df['total_amount'].mean()
# df['high_value_order'] = df['total_amount'] > average_amount
# df = df.sort_values(by='total_amount', ascending=False)
# df.head()

total = int(df['total_amount'].mean())

df['high_value_order'] = df['total_amount'].apply(lambda x: True if x > total else False)
df = df.sort_values(by='total_amount', ascending=False)
df.head()

```

```python
#6

df['average_rating']=df.groupby('country')['rating'].transform("mean")
df.head()
```

```python
#7

filtered = df.query("total_amount > 1000 and rating >4.5")
```

```python
#8
filtered['delivery_status'] = df['shipping_days'].apply(lambda x: 'delayed' if x > 7 else 'on_time')
filtered.head()
```

```python
filtered.to_csv("clean_orders_[eliezer_yakovson].csv", index=False, encoding="utf-8-sig")

```
